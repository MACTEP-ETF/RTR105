```python
print("Teilora rindas attēlošana funkcijai 'y=cos(x)*cos(x)' ASCII kodā:\n\n");

print("                   5  0  0\n");
print("  /|            ____________\n");
print(" / |            \\          |\n");
print("/  |             \\                  k+1       2*k      2*k-1\n");
print("   |              \\             (-1)     *   x     *  2\n");
print("   |   ------     /          ---------------------------------\n");
print("   |             /                       ( 2 * k ) !\n");
print("   |            /\n");
print("   |           /___________|\n");
print("                   k = 1\n");


print("\n\n\n\nRekurences attēlošana funkcijai 'y=cos(x)*cos(x)' ASCII kodā:\n\n");

print("           1\n");
print("       (-1)  * x * x * 4\n");
print("   --------------------------\n");
print("     (2 * k - 1) * (2 * k)\n\n\n\n");



k = 1 # īpašo mainīgo definēšana

print ("\nSveicināti!\n")
print ("Šī programma aprēķina funkcijas 'y=cos(x)*cos(x)' vērtību!\nDiemžēl, korekti skaitļi pēc Teilora rindas tikai pie 'x' intervālā [-17;17]\n")
x = float(input("\nLūdzu, ievadiet 'x' vērtību intervālā [-17;17]: "))
a = (pow(-1,k+1)*pow(x,2*k)*pow(2,2*k-1))/(1.*1*2)
S = a
while k < 500:
    k += 1
    a = a * (((-1)*x*x*4)/((2*k-1)*(2*k)))
    S = S + a
    if k == 25:
        b = a # saglabājam 25-to saskaitāmo
    if k == 75:
        z = a # saglabājam 75-to saskaitāmo
        
# attēlosim rezultātu
from math import cos
y = cos(x)*cos(x)
print ("\nAprēķins izmantojot funkciju:\ny=cos(%.2f)*cos(%.2f)=%.2f\n\n"%(x,x,y)) # izvadam gala rezultātu uz ekrana
print ("Aprēķins pēc Teilora rindas:\ny=%.2f\n\n"%(1-S))
print ("25-tais Teilora rindas saskaitāmais: %.e\n\n"%(b)) #izvadam konkrētu saskaitāmo uz ekrāna
print ("75-tais Teilora rindas saskaitāmais: %.e\n\n"%(z)) #izvadam konkrētu saskaitāmo uz ekrāna

# importēšu nepieciešamās bibliotēkas funkciju grafiku zīmēšanai 
from matplotlib import pyplot as plt 
import numpy

# izslēdzu jebkāda veida brīdinājumu paziņojumus
import warnings
warnings.filterwarnings("ignore")

# Ģenerējam funkciju un x vērtības, ja x ir diapazonā [-17;17]
x = []; y_funkcija = []; y_Teilor = []

for i in numpy.arange(-17, 17, 0.1):
    x.append(i)
    y_funkcija.append(cos(i)*cos(i))
    
    #aprēķins Teilora rindas funkcijai
    a = (pow(-1,k+1)*pow(i,2*k)*pow(2,2*k-1))/(1.*1*2)
    S = a
    while k < 500:
        k += 1
        a = a * (((-1)*i*i*4)/((2*k-1)*(2*k)))
        S = S + a
    y_Teilor.append(1-S)
    k = 1

print ("\n")

# Palielināšu grafiku un fontu izmērus
params = {'legend.fontsize': 15,
          'figure.figsize': (20, 10),
         'axes.labelsize': 25,
         'axes.titlesize':40,
         'xtick.labelsize':25,
         'ytick.labelsize':25}
plt.rcParams.update(params)

# Grafika zīmēšana pēc funkcijas un Teilora rindas
plt.plot(x, y_Teilor, label = "pēc Teilora rindas", linewidth = 10, color = 'green') 
plt.plot(x, y_funkcija, label = "pēc funkcijas cos(x)*cos(x)", linewidth = 3, color = 'orange') 
plt.title('Abu funkciju attēlojums vienā grafikā ar x vērtības diapazonu [-17;17]') 
plt.xlabel('x vērtība') 
plt.ylabel('f(x)') 
plt.legend(loc = 'lower right') 
plt.show() 

```
